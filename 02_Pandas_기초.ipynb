- 다중 조건을 주어 마스킹
    - 데이터프레임에서는 논리 연산자를 지원하지 않는다.
    - `or` 대신 `|` (Vertical bar)
        - 둘중에 하나가 1이면 1을 반환
    - `and` 대신 `&` (Ampersand)
        - 둘다 1이면 1을 반환
    - `not` 대신 `~` (tilde)
        - 1이면 0을 반환

```python
mask1 = df["target"] ==1
mask2 = df2["age"] < 20
df2.loc[mask1 & mask2].head()
```

- 여러 컬럼 추가하기

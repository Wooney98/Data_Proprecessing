```
- 고유한 고객번호로만 이루어진 데이터프레임을 만들어 features 변수에 넣으시오.
- 컬럼명은 고객번호로 하시오.
- 컬럼이 고객번호만 있는 데이터프레임을 만들라는 의미
- 중복 고객 번호가 있으면 안된다.
```

````python
arr = df["고객번호"].unique()
columns = ["고객번호"]
features = pd.DataFrame(arr,columns = columns)
features
````

```
groupby 메소드를 이용하여 고객별로 구매가격의 평균을 구하고 
데이터프레임 형식으로 정돈하여 구매가격의 컬럼명을 변경하여
features 변수의 데이터프레임과 left merge 하시오.
```

````python
tmp = df.groupby("고객번호")["구매가격"].mean().reset_index().rename(columns={"구매가격":"구매가격_평균"})
tmp
````

````python
features = pd.merge(features,tmp,how="left",on="고객번호")
features
````

```
groupby 메소드를 이용하여 고객별로 구매가격의 최대값과 대분류의 고유값개수(다양성)를 데이터프레임 형식으로 정돈하고
구매가격과 대분류의 컬럼명을 변경하여
features 변수의 데이터프레임과 left merge 하시오.
```

````python
agg_dict = {"구매가격":"max", "대분류":"nunique"}

tmp = df.groupby("고객번호").agg(agg_dict).reset_index().rename(columns={"구매가격":"최대가격","대분류":"대분류다양성"})
tmp
````

````python
features = pd.merge(features,tmp,how="left",on="고객번호")
features
````

```
groupby 메소드를 이용하여 고객별로 구매가격의 최소값과, 총구매액,
그리고 지점의 고유값 개수와 최빈값을 구하시오
구하는 과정에서 집계 컬럼에 대한 컬럼명도 정의하시오.
데이터 프레임 형식으로 정돈하여 
features 변수의 데이터프레임과 left merge 하시오.
```

````python
df.head(5)
````

````python
agg_dict = {
    "구매가격":[
        ("소비최소값",lambda x:x[x>0].min()), # 음수 제외 마스킹
        ("총구매액","sum")
        ],
     "지점":[
         ("고유값개수","nunique"),
         ("최빈값",lambda x:x.mode())
         ]
    }

tmp = df.groupby("고객번호").agg(agg_dict)
tmp.columns = tmp.columns.droplevel()
tmp = tmp.reset_index()
tmp
````

````python
pd.merge(features,tmp,how="left",on="고객번호")
````
